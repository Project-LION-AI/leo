#### template 2:

```python
from langchain import PromptTemplate, FewShotPromptTemplate

# First, create the list of few shot examples.
examples = [
    {"message": intro1, "class": "intro"},
    {"message": intro2, "class": "not intro"},
    {"message": "Welcome! Thank you for joining our community!", "class": "not intro"}
]

# Next, we specify the template to format the examples we have provided.
# We use the `PromptTemplate` class for this.
example_formatter_template = """
message: {message}
class: {class}\n
"""
example_prompt = PromptTemplate(
    input_variables=["message", "class"],
    template=example_formatter_template,
)

# Finally, we create the `FewShotPromptTemplate` object.
few_shot_prompt = FewShotPromptTemplate(
    # These are the examples we want to insert into the prompt.
    examples=examples,
    # This is how we want to format the examples when we insert them into the prompt.
    example_prompt=example_prompt,
    # The prefix is some text that goes before the examples in the prompt.
    # Usually, this consists of intructions.
    prefix="Predict whether or not a message is an introduction",
    # The suffix is some text that goes after the examples in the prompt.
    # Usually, this is where the user input will go
    suffix="message: {input}\nclass:",
    # The input variables are the variables that the overall prompt expects.
    input_variables=["input"],
    # The example_separator is the string we will use to join the prefix, examples, and suffix together with.
    example_separator="\n\n",
)

# We can now generate a prompt using the `format` method.
#print(few_shot_prompt.format(input=intro3))
```

```python
from langchain.prompts.example_selector import LengthBasedExampleSelector


# These are a lot of examples of a pretend task of creating antonyms.
examples = [
    {"message": intro1, "class": "intro"},
    {"message": intro2, "class": "not intro"},
    {"message": "Welcome! Thank you for joining our community!", "class": "not intro"},
    {"message": "Hi, my name is carl! I love what you all are doing. Glad to be here.", "class": "intro"},
    {"message": "Great to have you!", "class": "not intro"},
]

# We'll use the `LengthBasedExampleSelector` to select the examples.
example_selector = LengthBasedExampleSelector(
    # These are the examples is has available to choose from.
    examples=examples, 
    # This is the PromptTemplate being used to format the examples.
    example_prompt=example_prompt, 
    # This is the maximum length that the formatted examples should be.
    # Length is measured by the get_text_length function below.
    max_length=200,
)

# We can now use the `example_selector` to create a `FewShotPromptTemplate`.
dynamic_prompt = FewShotPromptTemplate(
    # We provide an ExampleSelector instead of examples.
    example_selector=example_selector,
    example_prompt=example_prompt,
    prefix="Predict whether or not a message is an introduction.",
    suffix="message: {input}\nclass:",
    input_variables=["input"],
    example_separator="\n\n",
)

# We can now generate a prompt using the `format` method.
my_prompt = dynamic_prompt.format(input=intro3)

print(my_prompt)
```

### Process for predicting intros

```python
# intros list to dataframe
import pandas as pd
data = pd.DataFrame(intros[103:120])
data = data.rename(columns={0: "message"})
# # correcting the above line
classes = ["intro", "not intro", "not intro", "not intro", "not intro", "not intro", "not intro", "not intro", "not intro", 
           "intro", "not intro", "not intro", "intro", "not intro", "not intro", "not intro", "intro"]
data["class"] = classes
data

# split into train and test
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.5, random_state=42)
train
```

```python
# define example set

# These are a lot of examples of a pretend task of creating antonyms.
examples = [
    {"message": intro1, "class": "intro"},
    {"message": intro2, "class": "intro"},
    {"message": "Welcome! Thank you for joining our community!", "class": "not intro"},
    {"message": "Hi, my name is carl! I love what you all are doing. Glad to be here.", "class": "intro"},
    {"message": "Great to have you!", "class": "not intro"},
]

# add the training set to the examples dict
for i in range(len(train)):
    examples.append({"message": train.iloc[i, 0], "class": train.iloc[i, 1]})
```

```python
# define the example selector
from langchain.prompts.example_selector import LengthBasedExampleSelector

# We'll use the `LengthBasedExampleSelector` to select the examples.
example_selector = LengthBasedExampleSelector(
    # These are the examples is has available to choose from.
    examples=examples, 
    # This is the PromptTemplate being used to format the examples.
    example_prompt=example_prompt, 
    # This is the maximum length that the formatted examples should be.
    # Length is measured by the get_text_length function below.
    max_length=300,
)

# We can now use the `example_selector` to create a `FewShotPromptTemplate`.
dynamic_prompt = FewShotPromptTemplate(
    # We provide an ExampleSelector instead of examples.
    example_selector=example_selector,
    example_prompt=example_prompt,
    prefix="Predict whether or not a message is an introduction.",
    suffix="message: {input}\nclass:",
    input_variables=["input"],
    example_separator="\n",
)
```

```python
# select a random item from the intros list
import random
sample_intro = random.choice(intros)

# We can now generate a prompt using the `format` method.
my_prompt = dynamic_prompt.format(input=sample_intro)
print(my_prompt)
```

```python
# Initialize an empty dictionary
results_dict = {}

for message in test.iterrows():
    my_prompt = dynamic_prompt.format(input=message[1][0])
    llm = OpenAI(model_name="text-davinci-003", n=2, best_of=2, temperature=0.9)
    result = llm(my_prompt)
    
    # Store the input message and the result in the dictionary directly
    results_dict[message[1][0]] = result

#print(results_dict)
```

```python
# now join on the predictions onto the test set
test["predictions"] = test["message"].map(results_dict)
test.sample()

# lets convert the classes to 0 and 1
test["class"] = test["class"].map({"intro": 1, "not intro": 0})
test["predictions"] = test["predictions"].str.strip().map({"intro": 1, "not intro": 0}) # there are white spaces in the predictions
# now create an accuracy column
test["accuracy"] = test["class"] == test["predictions"]
test.sample(5)
```

```python
# calculate the accuracy
test["accuracy"].mean()
```
